```python
nlp = spacy.load("en_core_web_sm")
```

```python
text_list = df["text"].tolist()
text_list
```

```python
aspects = []
for sentence in text_list:
    doc = nlp(sentence)
    descriptive_term = ''
    target = ''
    for token in doc:
        if token.dep_ == 'nsubj' and token.pos_ == 'NOUN':
            target = token.text
        if token.pos_ == 'ADJ':
            prepend = ''
            for child in token.children:
                if child.pos_ != 'ADV':
                    continue
                prepend += child.text + ' '
            descriptive_term = prepend + token.text
    aspects.append({'aspect': target,'description': descriptive_term})
print(aspects)
```

```python
#!pip install -U textblob
!python -m textblob.download_corpora
```

```python
from textblob import TextBlob
for aspect in aspects:
    polarity,subjectivity=TextBlob(aspect['description']).sentiment
    aspect['polarity'] = polarity
    aspect['subjectivity']=subjectivity
print(aspects)
```

```python
df_aspects=pd.DataFrame.from_dict(aspects)
```

```python
df_aspects
```

```python
import pandas
df2=pandas.merge(df, df_aspects, left_index=True, right_index=True)
df2.head()
```

```python
def senti(data):
    if data['polarity'] >= 0.05:
        val = 1
    elif data['polarity'] <= -0.05:
        val = -1
    else:
        val = 0
    return val
```

```python
df2['Sentiment'] = df2.apply(senti,axis=1)
```

# Count Vectorizer
I will use a count vectorizer to vectorize the text data in the review column (training feature for this project) and then use three different classification models from scikit-learn models. After that, to evaluate the model on this dataset find out the accuracy, confusion matrix, true positive rates, and true negative rates. Here are the steps.

* The first step is to split the dataset into training sets and testing sets.
* Vectorize the input feature that is out review column (both training and testing data)
* import the model from scikit learn library.
* Find the accuracy score
* find the true positive and true negative rates.

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
ctmTr = cv.fit_transform(X_train)
X_test_dtm = cv.transform(X_test)
```

# LogisticRegression

```python
import warnings
warnings.filterwarnings("ignore")
from sklearn.linear_model import LogisticRegression
#Training the model
lr = LogisticRegression()
lr.fit(ctmTr, y_train)

#Accuracy score
lr_score = lr.score(X_test_dtm, y_test)
print("Results for Logistic Regression with CountVectorizer")
print(lr_score)

#Predicting the labels for test data
y_pred_lr = lr.predict(X_test_dtm)

from sklearn.metrics import confusion_matrix
#Confusion matrix
cm_lr = confusion_matrix(y_test, y_pred_lr)
TN = cm_lr[0][0]
FN = cm_lr[1][0]
TP = cm_lr[1][1]
FP = cm_lr[0][1]
print('true positive,False negative,True positive,False positive')
print(TN,FN,TP,FP)

#True positive and true negative rates
print('True positive and true negative rates')
tpr_lr = round(TP/(TP + FN), 4)
tnr_lr = round(TP/(TN+FP), 4)
print(tpr_lr, tnr_lr)

#Evaluate the model using other performace metrics
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred_lr))
```
